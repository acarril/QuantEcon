# Vectores

Un vector de longitud $n$ es simplemente una secuencia (o *array*, o tupla) de $n$ números, escrito como
$$ x = [x_1, x_2, \ldots, x_n]. $$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Comenzar figura con dos subfiguras
fig, ax = plt.subplots(figsize=(10, 8))
# Set the axes through the origin
for spine in ['left', 'bottom']:
    ax.spines[spine].set_position('zero')
for spine in ['right', 'top']:
    ax.spines[spine].set_color('none')

ax.set(xlim=(-5, 5), ylim=(-5, 5))
ax.grid()
vecs = ((2, 4), (-3, 3), (-4, -3.5))
for v in vecs:
    ax.annotate('', xy=v, xytext=(0, 0),
                arrowprops=dict(facecolor='blue',
                shrink=0,
                alpha=0.7,
                width=0.5))
    ax.text(1.1 * v[0], 1.1 * v[1], str(v))
plt.show()
```

## Operaciones con vectores

Dos de las operaciones más comunes con vectores son la suma y la multiplicación por un escalar.

Por definición, al sumar dos vectores sumamos elemento por elemento:

$$ 8+2 $$

```python
x = np.array([1,2,3])
y = np.array([4,5,6])
x + y
```

```python
z = np.ones(3) # Crear un vector de 1s con n=3
y + z
```

```python
4 * z
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
# Set the axes through the origin
for spine in ['left', 'bottom']:
    ax.spines[spine].set_position('zero')
for spine in ['right', 'top']:
    ax.spines[spine].set_color('none')

ax.set(xlim=(-5, 5), ylim=(-5, 5))
x = (2, 2)
ax.annotate('', xy=x, xytext=(0, 0),
            arrowprops=dict(facecolor='blue',
            shrink=0,
            alpha=1,
            width=0.5))
ax.text(x[0] + 0.4, x[1] - 0.2, '$x$', fontsize='16')


scalars = (-1.5, 2)
x = np.array(x)

for s in scalars:
    v = s * x
    ax.annotate('', xy=v, xytext=(0, 0),
                arrowprops=dict(facecolor='red',
                shrink=0,
                alpha=0.5,
                width=0.5))
    ax.text(v[0] + 0.4, v[1] - 0.2, f'${s} x$', fontsize='16')
plt.show()
```

# Producto Interno y Norma

El **producto interno** (o **producto escalar**, o **producto punto**) de los vectores $x,y \in \mathbb{R}^n$ se define como
$$ x'y := \sum_{i=1}^n x_i y_i. $$
Decimos que dos vectores son ortogonales si su producto interno es cero.

La **norma** de un vector $x$ puede interpretarse como su "longitud" (ie. su distancia del vector 0), y se define como
$$ \Vert x \Vert := \sqrt{x'x} := \left( \sum_{i=1}^n x_i^2 \right)^{1/2}. $$

```python
np.sum(x*y) # Forma manual (multiplicar cada término y luego sumar)
np.dot(x,y) # Función para cálculo directo
```

```python
# Norma
np.sqrt(np.sum(x**2)) # Método manual
np.linalg.norm(x) # Función
```
